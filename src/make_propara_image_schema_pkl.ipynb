```python
import json
import pandas as pd
```

```python
#path = '/nfs/research/regan/data_propara/data/corenlp_propara_v7.json'

path = '/nfs/research/regan/data_propara/data/corenlp_propara_v9.json'
```

```python
with open(path, 'r') as f:
    data = json.load(f)
len(data)
```

```python
data.keys()
```

```python
all_items = []

for k, v in data.items():
    
    if 'metadata' not in k:
    
        if v["fd1"] != '':
            
            if len(v['fd3'])>0:
                fd3 = v['fd3'].lower()
            else:
                fd3 = 'none'

            item = {"sentence": v["example"], 
                    "syntax":v["syntax"], 
                    "fd1":v["fd1"].lower(), 
                    "fd2":v["fd2"].lower(), 
                    "fd3":fd3, 
                    "verb":v["verb"].split('.')[0]}
            print(k, item)
            print()
            all_items.append(item)
    
all_items[1]
```

```python
len(all_items)
```

```python
df = pd.DataFrame(all_items)
```
